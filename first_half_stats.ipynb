```python
#import packages
import pandas as pd
from pybaseball import pitching_stats
```

```python
#Gather all qualified pitchers from 2000 to 2019
qualified_pitchers = pitching_stats(2000, 2019)

qualified_pitchers
```

```python
#Get every pitcher that appears in the dataframe
unique_pitchers_arr = qualified_pitchers.Name.unique()
unique_pitchers = pd.DataFrame(unique_pitchers_arr, columns= ['Name'])

unique_pitchers
```
