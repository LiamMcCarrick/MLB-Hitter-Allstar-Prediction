```python
#import packages
import pandas as pd
from pybaseball import batting_stats
```

# Replace Special Characters

```python
#read in first half stats to df
first_half_00_19 = pd.read_csv("Model_Data\\first_half_stats.csv")
first_half_21 = pd.read_csv("Model_Data\\first_half_stats_2021.csv")
```

```python
first_half_00_19
```

```python
# replace special characters
cols = first_half_00_19.select_dtypes(include=[object]).columns
first_half_00_19[cols] = first_half_00_19[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))

cols21 = first_half_21.select_dtypes(include=[object]).columns
first_half_21[cols21] = first_half_21[cols21].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))

first_half_21.to_csv("Model_Data\\first_half_stats_2021.csv")
first_half_00_19.to_csv("Model_Data\\first_half_stats.csv")
```
